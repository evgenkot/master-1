```python
import sys
import os

sys.path.append('../task1')

import mnist_loader
import network
```

```python
import time

training_data, validation_data, test_data = mnist_loader.load_data_wrapper('../task1/mnist.pkl.gz')
validation_data_len = len(list(validation_data))

results = []

configurations = [
    # Референсная конфигурация
    ([784, 30, 10], 15, 10, 3.0),
    
    # Конфигурация 1: Увеличение числа нейронов в скрытом слое
    ([784, 60, 10], 15, 10, 3.0),
    
    # Конфигурация 2: Уменьшение числа нейронов в скрытом слое
    ([784, 15, 10], 15, 10, 3.0),
    
    # Конфигурация 3: Увеличение числа эпох
    ([784, 30, 10], 30, 10, 3.0),
    
    # Конфигурация 4: Уменьшение числа эпох
    ([784, 30, 10], 7, 10, 3.0),
    
    # Конфигурация 5: Увеличение размера мини-батча
    ([784, 30, 10], 15, 20, 3.0),
    
    # Конфигурация 6: Уменьшение размера мини-батча
    ([784, 30, 10], 15, 5, 3.0),
    
    # Конфигурация 7: Увеличение скорости обучения
    ([784, 30, 10], 15, 10, 5.0),
    
    # Конфигурация 8: Уменьшение скорости обучения
    ([784, 30, 10], 15, 10, 0.5),

    # Конфигурация 9: Добавление скрытого слоя
    ([784, 100, 30, 10], 15, 10, 3.0),
    
    # Конфигурация 10: Удаление скрытого слоя
    ([784, 10], 15, 10, 3.0)
]

# Обучение и тестирование каждой конфигурации
for topology, epochs, mini_batch_size, learning_rate in configurations:
    training_data, validation_data, test_data = mnist_loader.load_data_wrapper('../task1/mnist.pkl.gz')
    net = network.Network(topology)
    print(f"Training network with topology: {topology}, epochs: {epochs}, mini_batch_size: {mini_batch_size}, learning_rate: {learning_rate}")
    start_time = time.time()
    net.SGD(training_data, epochs, mini_batch_size, learning_rate, test_data=test_data)
    end_time = time.time()
    training_time = end_time - start_time

    correct_answers = net.evaluate(validation_data)
    accuracy = correct_answers / validation_data_len
    results.append((topology, accuracy, training_time))
    print(f"Result Topology: {topology}, Accuracy: {accuracy:.4f}, Training Time: {training_time:.2f} seconds")

for topology, accuracy, training_time in results:
    print(f"Topology: {topology}, Accuracy: {accuracy:.4f}, Training Time: {training_time:.2f} seconds")
```


## 1. Параметры метода обучения
Качество обучения нейронной сети зависит от нескольких ключевых параметров:

- **Количество эпох**: Увеличение числа эпох может улучшить качество обучения, но также может привести к переобучению, если сеть будет слишком долго обучаться на одном и том же наборе данных.
  
- **Размер мини-батча**: Большие мини-батчи могут ускорить обучение, но могут привести к менее точным градиентам, в то время как маленькие мини-батчи обеспечивают более точные обновления, но требуют больше времени на обучение.

- **Скорость обучения**: Высокая скорость обучения может привести к быстрому обучению, но также может вызвать нестабильность и пропуск минимума функции потерь. Низкая скорость обучения обеспечивает стабильность, но может замедлить процесс.

## 2. Влияние топологии сети
Топология сети, включая количество слоев и количество нейронов в каждом слое, существенно влияет на качество обучения:

- **Количество скрытых слоев**: Увеличение числа скрытых слоев может повысить способность сети к обучению сложным функциям, но также увеличивает риск переобучения.

- **Количество нейронов в скрытых слоях**: Большее количество нейронов может улучшить качество распознавания, но также увеличивает вычислительные затраты и время обучения.
